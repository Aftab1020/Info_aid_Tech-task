### Preprocessing

```python
# finding missing values

df.isnull().sum()
```

```python
# dropping overview missing rows

df.dropna(inplace=True)
```

#### Creating function to convert columns dictionary to list

```python
def transfer(items):
    bucket = []
    for i in ast.literal_eval(items):
        bucket.append(i['name'])
    return bucket
    
```

Converting genres dictionary to list

```python
df['genres'] = df['genres'].apply(transfer)
```

```python
df.head()
```

 Converting keywards dictionary to list

```python
df['keywords'] = df['keywords'].apply(transfer)
```

```python
df.head()
```

```python
df['overview'][0]
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
def transfer1(items):
    bucket = []
    counter = 0
    for i in ast.literal_eval(items):
        if counter != 3:
            bucket.append(i['name'])
            counter += 1
        else:
            break
    return bucket
    
```

```python
df['cast'] = df['cast'].apply(transfer1)
```

```python
df.head()
```

```python
def director_name(items):
    bucket = []
    for i in ast.literal_eval(items):
        if i['job'] == 'Director':
            bucket.append(i['name'])
            break
    return bucket
```

```python
df['crew'] = df['crew'].apply(director_name)
```

```python
df.head()
```

#### Removing spaces from data

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.head()
```

```python
data = df[['movie_id','title','tags']]
```

```python
data.head()
```

```python
# convert tags list to string

data['tags'] = data['tags'].apply(lambda x:" ".join(x))
```

```python
data.head()
```

```python
# convert tags data to lowercase
data['tags'] = data['tags'].apply(lambda x:x.lower())
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(data['tags']).toarray()
```

```python
vector[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1] )[1:10]
```

```python
def recommend(movie):
    index = data[data['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(data.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
